# Тульская Дозиметрия ☠️

Дан текстовый файл, содержащий данные по измерениям радиационного излучения в Тульской области с января по сентябрь. Каждое измерение имеет свой идентификационный десятизначный цифровой код, значение излучения (микрозиверт/час), широта и долгота измерения, Горизонтальная точность — радиус, в котором значение находится с 68% вероятностью, тип маркера и дата измерения.

Значения даты даны в формате число+точка+три первые буквы месяца, но часть из них введена некорректно: у некоторых значений вместо первых букв месяца стоит 0, у других строка 'not provided'. Так как значения, не привязанные по времени, не несут смысловой нагрузки в данных, их требуется удалить.

Также, по требованию лаборатории, значения радиуса точности должны не превышать 25.



```python
with open('data/tula_doze.txt', encoding = 'utf-8') as f:
    t_list = f.readlines()   
    
names = t_list[0].split() # создаем список с названиями столбцов

t_dict = {} 

for i in range(1, len(t_list)):
    el = t_list[i].split()
    cur_dict = {}         # создаем текущий вложенный список, который станет значением ключа 
    for j in range(1, len(names)):
        cur_dict[names[j]] = el[j]      # ключи текущего списка — названия столбцов, значения — соответствующие значения параметров для данного образца
    t_dict[el[0]]= cur_dict
      
# t_dict — словарь, ключи которого это номера измерений, а значения представлены в виде вложенных словарей

t_dict_unsorted = t_dict

for key in t_dict.copy():
    if len(t_dict[key]['date'])!=6 or int(t_dict[key]['horizontal_accuracy']) > 25: #если название месяца указано корректно, в нем 6 символов
            t_dict.pop(key)
    

print(t_dict)
        
        
        
```

# Распределение по месяцам 

На этом этапе осуществляется сортировка значений измерений по месяцам в созданном словаре, а также поиск среднего значения радиационного излучения за каждый месяц.

```python
jan = [] # создаем списки за каждый месяц
feb = []
mar = []
apr = []
may = []
jun = []
jul = []
aug = []
sep = []
jan_sum = 0 # для поиска среднего можно просуммировать все значения за месяц и после разделить их на количество осуществленных измерений
feb_sum = 0 # задаем изначальную сумму
mar_sum = 0
apr_sum = 0
may_sum = 0
jun_sum = 0
jul_sum = 0
aug_sum = 0
sep_sum = 0
jan_counter = 0 # задаем количество измерений в каждом месяце, равное нулю
feb_counter = 0
mar_counter = 0
apr_counter = 0
may_counter = 0
jun_counter = 0
jul_counter = 0
aug_counter = 0
sep_counter = 0
for key in t_dict:
    if t_dict[key]['date'][3:]=='янв': # устанавливаем, что  после date необходимо начинать считывание с 3 символа, чтобы извлечь название месяца
        jan.append(float(t_dict[key]['doserate_(uSv/h)'])) # добавляем значение радиационного излучения к сумме
        jan_sum += float(t_dict[key]['doserate_(uSv/h)']) # добавляем единицу к количеству измерений
        jan_counter+=1
    elif t_dict[key]['date'][3:]=='фев':
        feb.append(float(t_dict[key]['doserate_(uSv/h)']))
        feb_sum += float(t_dict[key]['doserate_(uSv/h)'])
        feb_counter+=1
    elif t_dict[key]['date'][3:]=='мар':
        mar.append(float(t_dict[key]['doserate_(uSv/h)']))
        mar_sum += float(t_dict[key]['doserate_(uSv/h)'])
        mar_counter+=1
    elif t_dict[key]['date'][3:]=='апр':
        apr.append(float(t_dict[key]['doserate_(uSv/h)']))
        apr_sum += float(t_dict[key]['doserate_(uSv/h)'])
        apr_counter+=1
    elif t_dict[key]['date'][3:]=='май':
        may.append(float(t_dict[key]['doserate_(uSv/h)']))
        may_sum += float(t_dict[key]['doserate_(uSv/h)'])
        may_counter+=1
    elif t_dict[key]['date'][3:]=='июн':
        jun.append(float(t_dict[key]['doserate_(uSv/h)']))
        jun_sum += float(t_dict[key]['doserate_(uSv/h)'])
        jun_counter+=1
    elif t_dict[key]['date'][3:]=='июл':
        jul.append(float(t_dict[key]['doserate_(uSv/h)']))
        jul_sum += float(t_dict[key]['doserate_(uSv/h)'])
        jul_counter+=1
    elif t_dict[key]['date'][3:]=='авг':
        aug.append(float(t_dict[key]['doserate_(uSv/h)']))
        aug_sum += float(t_dict[key]['doserate_(uSv/h)'])
        aug_counter+=1
    else:
        sep.append(float(t_dict[key]['doserate_(uSv/h)']))
        sep_sum += float(t_dict[key]['doserate_(uSv/h)'])
        sep_counter+=1
dict_mean= {} # создаем словарь со средними значениями по месяцам, равными частному сумм измерений радиации и количеству измерений
dict_mean['January'] = jan_sum/jan_counter
dict_mean['February'] = feb_sum/feb_counter
dict_mean['March'] = mar_sum/mar_counter
dict_mean['April'] = apr_sum/apr_counter
dict_mean['May'] = may_sum/may_counter
dict_mean['June'] = jun_sum/jun_counter
dict_mean['July'] = jul_sum/jul_counter
dict_mean['August'] = aug_sum/aug_counter
dict_mean['September'] = sep_sum/sep_counter

print(dict_mean)
```
